#  Objetivos
- buscar ações de forma automática no site do Yahoo Finanças
- gerar as análises solicitadas pelo gestor
- enviar um e mail automatizado com resultados das analise para o gestor


# Buscar Dados


```python
import yfinance
```

```python
ticker = input("Digite a sigla da ação desejada: ")
periodo_acao = input("Digite o período da ação: ")
dados = yfinance.Ticker(ticker).history(periodo_acao)
fechamento = dados.Close
fechamento.plot()
                                
```

# Gerar as análises solicitadas pelo gestor

- Análises dos últimos 6 meses
- Cotação Máxima
- Cotação Mínima
- Cotação Atual

```python
cotacao_maxima = fechamento.max()
cotacao_minima = fechamento.min()
cotacao_atual = fechamento[-1]

```

# Enviar os dados por e mail automaticamente

```python
import pyautogui
import pyperclip
```

```python
pyautogui.PAUSE = 3

pyautogui.hotkey("ctrl", "t")
pyperclip.copy("www.gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")
pyautogui.click(x=84, y=218)

pyperclip.copy("fernandogazoli@hotmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

pyperclip.copy("Análise de Ações")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")


mensagem = f"""

Prezado,

Seguem as cotações da empresa {ticker} conforme solicitado dos últimos 6 meses:

Cotação Mínima R$ {round(cotacao_minima, 2)}
Cotação Máxima R$ {round(cotacao_maxima, 2)}
Cotação Atual R$ {round(cotacao_atual, 2)}

Qualquer dúvida, estou à disposição

Atte.

"""
pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")

pyautogui.hotkey("enter")
pyautogui.click(x=1129, y=983)


```
